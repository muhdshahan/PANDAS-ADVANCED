PERFORMING NORMALIZATION [ MIN - MAX SCALING]

```python
import pandas as pd
data = pd.read_csv('Titanic-Dataset.csv')
data
```

```python
#sklearn.preprocessing is the library that contains the function minmaxscaler
from sklearn.preprocessing import MinMaxScaler

# minmaxscaler function transforms data values to specific range (default is between 0 and 1)
scaling = MinMaxScaler()
#fit_transform function fits the age column to the range , knows the max and min values and sets accordingly
data[['Age','Fare']] = scaling.fit_transform(data[['Age','Fare']])
data
```

FEATURE SCALING USING STANDARDIZATION [ Z-SCORE NORMALIZATION ] TECHNIQUE ( AVOIDS OUTLIERS ALSO )

```python
import pandas as pd
data1 = pd.read_csv('Titanic-Dataset.csv')
```

```python
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()
#SCALES VALUES WITH MEAN 0 AND STD 1
data1[['Age','Fare']] = scale.fit_transform(data1[['Age','Fare']])
data1
```

FEATURE SCALING WITH MAX ABS SCALING, RHOBUST SCALING

```python
import pandas as pd
data2 = pd.read_csv('Titanic-Dataset.csv')
```

```python
from sklearn.preprocessing import MaxAbsScaler 
scaler = MaxAbsScaler()
#SCALES VALUS WITH -1 TO 1 IF DATA CONTAINS BOTH +VE AND -VE , 0 TO 1 IF DATA ONLY  +VE ,0 TO -1 IF DATA ONLY  -VE
data2[['Age','Fare']] = scaler.fit_transform(data2[['Age','Fare']])
data2
```

```python
data2.to_csv('Titanic-MaxAbs feature scaling.csv')
```
