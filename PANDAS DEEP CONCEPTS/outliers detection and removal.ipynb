```python
import pandas as pd

data = pd.read_csv('Titanic-Dataset.csv')
data
```

DETECTION OF OUTLIERS USING STANDARD DEVIATION

```python
mean = data['Fare'].mean()
std = data['Fare'].std()

# setting threshold
threshold = 2

upper_bound = mean + ( threshold * std)
lower_bound = mean - ( threshold * std)

data['std outlier'] = ( (data['Fare'] < lower_bound) | (data['Fare'] > upper_bound))
data
```

```python
data.to_csv('Titanic-outliers detected.csv')
```

OUTLIERS DETECTION USING Z-SCORE

```python
details = pd.read_csv('Titanic-Dataset.csv')

details['z-score'] = ( details['Fare'] - mean) / std
details['z-score outliers'] = details['z-score'].abs() > 3
details
```

```python
details.to_csv('Titanic-zscore outlier detection.csv')
```

OUTLIERS DETECTION USING PERCENTILE METHOD AND THEN CAPPING TECHNIQUE  

```python
passengers = pd.read_csv('Titanic-Dataset.csv')
passengers
```

```python
upper = passengers['Fare'].quantile(0.95)
lower = passengers['Fare'].quantile(0.05)

passengers['outliers'] = ( (passengers['Fare'] > upper) | (passengers['Fare'] < lower) )
passengers
```

```python
passengers.to_csv('Titanic-percentile outlier detection.csv')
```

REMOVING OUTLIERS WITH THE PERCENTILE METHOD ITSELF

```python
passengers['Fare'] = passengers['Fare'].apply( lambda x : upper if x > upper else x)

passengers['Fare'] = passengers['Fare'].apply( lambda x : lower if x < lower else x)
passengers.to_csv('Titanic-outlier removed.csv')
passengers
```

```python
# or

passengers['Fare'] = passengers['Fare'].clip( lower , upper )
passengers.to_csv('Titanic-outlier removal capping with clip.csv')
```
