PERFORM SPLITTING ( SPLIT TO TRAIN SETS AND TEST SETS ) OF DATASET CONTAINING DATA OF 50,000 HOUSES 

```python
import pandas as pd
dataset = pd.read_csv('housing_price_dataset.csv')
dataset
```

```python
from sklearn.model_selection import train_test_split

# x= features , y = target
x = dataset[['SquareFeet' , 'Bedrooms' , 'Bathrooms' , 'Neighborhood' , 'YearBuilt']]
y = dataset['Price']
#normal splitting to test and train data
x_train  , x_test , y_train  , y_test = train_test_split(x , y , random_state= 0)
#random_state performs the function of seed reproducibility so the order of data will be set once
x_train
```

```python
x_test
```

```python
#splitting  test to 20% and train 80%
x_train  , x_test , y_train  , y_test = train_test_split(x , y , random_state= 0 , test_size=0.2)
#instead of test_size there also option to set train_size
x_train
```

```python
x_test
```

SPLITTING THE TRAINED DATA AGAIN TO TRAIN AND VALIDATION SETS

```python
from sklearn.model_selection import train_test_split
import pandas as pd

split_data = pd.read_csv('housing_price_dataset.csv')

x = split_data[['SquareFeet' , 'Bedrooms' , 'Bathrooms' , 'Neighborhood' , 'YearBuilt']]
y = split_data['Price']

#splitting to train_val ( train and validation) and test data
x_train_val , x_test , y_train_val , y_test = train_test_split(x , y , test_size=0.2 , random_state=3)

#again splitting the train_val data to train and validation data
x_train , x_val  , y_train , y_val = train_test_split(x_train_val , y_train_val , test_size=0.2 , random_state=3)
x_train
```
